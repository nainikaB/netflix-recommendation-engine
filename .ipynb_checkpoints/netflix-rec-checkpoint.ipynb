```python
import pandas as pd
import numpy as np
```

```python
netflix = pd.read_csv('netflix_titles.csv')
netflix.head()
```

```python
netflix.shape
```

```python
netflix = netflix[['show_id',  'title', 'type', 'director', 'cast', 'release_year', 'rating', 'listed_in', 'description']]
```

```python
netflix.info()
```

```python
netflix.dropna(inplace = True)
netflix.shape
```

```python
netflix['type'] = netflix['type'].str.replace(" ", "")
netflix['director'] = netflix['director'].str.replace(" ", "")
netflix['cast'] = netflix['cast'].str.replace(" ", "")
netflix['listed_in'] = netflix['listed_in'].str.replace(" ", "")
netflix['description'] = netflix['description'].str.replace(" ",",")
netflix.head()
```

```python
netflix['tags'] =  netflix['type'] + " ," + netflix['director'] + "," + netflix['cast'] + "," +  netflix['rating']+ ","   +  netflix['listed_in']+ ","   +  netflix['description'] 
netflix.head()
```

```python
final_data = netflix[['show_id', 'title', 'tags']]
final_data.head()
```

```python
final_data['tags'] = final_data['tags'].str.replace(","," ")
```

```python
final_data.head()
```

```python
final_data['tags'] = final_data['tags'].apply(lambda x:x.lower())
```

```python
final_data.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    li = []
    for i in text.split():
        li.append(ps.stem(i))
        
    return " ".join(li)
```

```python
final_data['tags'] = final_data['tags'].apply(stem)
```

```python
final_data['tags']
```

```python
final_data.tags.str.split(expand=True).stack().value_counts()[:5000]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(final_data['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = final_data[final_data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final_data.iloc[i[0]].title)
```

```python
recommend('Midnight Mass')
```
